```python
#calling the necessary libraries

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

# Now we will get the url of the site and will store all the pages of the site in a variable

```python
url = 'https://www.imdb.com/list/ls068082370/'
all=[]
while(url != None):
    all.append(url)
    page=requests.get(url)
    soup = BeautifulSoup(page.text,'html.parser')
    next=soup.find_all(class_="flat-button lister-page-next next-page")
    if (len(next)==0):
        url= None
    else:
        next="https://www.imdb.com" + next[0].get('href')
        url=next
all
```

# We will now start working with the contents of the pages now

```python
#while storing the information of directors we found that the directors and stars are stored together.
#Thus we have created a function, that will store the names of the directors and stars separately.


def d_a(d):
    d=d.replace('\n',"")
    d=d.replace('|',"")
    d=d.split('Stars:')
    d[0]=d[0].replace("Directors:","")
    d[0]=d[0].replace("Director:","")
    for i in range(10):
        d[0]=d[0].replace("  "," ")
    director=d[0][1:]
    stars=d[1]
    stars=stars.replace(":","")
    
    return director,stars
```

```python
#The same problem has been found while working with the gross and votes.
#Similarly, we have created this function to separate the gross and the votes value

def v_g(v):
    vg_list=[]
    for i in v:
        vg_list.append(i.text)
    if(len(v)==2):
        votes=vg_list[0]
        gross=vg_list[1]
    else:
        votes=vg_list[0]
        gross=None
    
    return votes,gross
```

# Now that we have done separating with the different contents of the movies. We will pass all those informations in a final array.

```python
final = []
for i in all:
    page=requests.get(i)
    soup = BeautifulSoup(page.text,'html.parser')
    for j in soup.find_all(class_='lister-item-content'):
        ids =int(j.find('span',{"class":'lister-item-index unbold text-primary'}).text[:-1])
        movie=j.find('a').text
        year=j.find('span',{'class':'lister-item-year text-muted unbold'}).text[1:5]
        genre=j.find('span',{'class':'genre'}).text[1:]
        time = j.find('span',{'class':'runtime'}).text
        imdb=j.find('span',{'class':'ipl-rating-star__rating'}).text
        rating=j.find('span',{'class':'ipl-rating-star__rating'}).text
        summary=j.find_all('p')[1].text[5:]
        director,stars=d_a(j.find_all('p',{'class':'text-muted text-small'})[1].text)
        votes,gross=v_g(j.find_all('span',{'name':'nv'}))
        Add=[ids,movie,year,genre,time,imdb,rating,summary,director,stars,votes,gross]
        final.append(Add)
```

# Now that we have stored all the required informations in separate columns. We will name our columns accordingly.

```python
column=['ID','Movie name','Released Year','Genre','Time Length of the movie','IMDB','Ratings','Movie Summary','Directors','Stars','Votes','Gross']
```

# We will then convert the thing into a dataframe.

```python
data=pd.DataFrame(final,columns=column)
```
